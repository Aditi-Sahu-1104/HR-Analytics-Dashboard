```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1,'No' : 0})
```

```python
df['Attrition'].head()
```

```python
binary_cols = ['Gender', 'OverTime']
cat_cols= ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'Education']
num_cols = df.select_dtypes(include=['number']).columns
```

```python
num_cols = num_cols.drop('Attrition')
```

```python
from sklearn.preprocessing import FunctionTransformer

def binary_map(df):
    df = df.copy()
    df['Gender']   = df['Gender'].map({'Male':1, 'Female':0})
    df['OverTime'] = df['OverTime'].map({'Yes':1, 'No':0})
    return df

binary_transformer = FunctionTransformer(binary_map, validate=False)
```

```python
preprocess = ColumnTransformer(
    transformers = [
    ('numerical_transformation', StandardScaler(), num_cols),
    ('categorical_transformation', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols)
], remainder='passthrough')
```

```python
pipe = Pipeline([
    ('binary_map', binary_transformer),
    ('preprocess', preprocess),
    ('model',LogisticRegression(class_weight='balanced'))
])
```

```python
x = df.drop('Attrition', axis=1)
y=df['Attrition']
```

```python
final_model = grid_search.best_estimator_

lr = final_model.named_steps['model']

preprocessor = final_model.named_steps['preprocess']
```

```python
feature_names = preprocessor.get_feature_names_out()
```

```python
import pandas as pd
import numpy as np

coef_df = pd.DataFrame({
    'feature': feature_names,
    'coefficient': lr.coef_[0]
})

coef_df['odds_ratio'] = np.exp(coef_df['coefficient'])

coef_df = coef_df.sort_values(by='coefficient', ascending=False)

coef_df.reset_index(drop=True, inplace=True)
coef_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.barh(coef_df['feature'], coef_df['coefficient'])
plt.xlabel("Coefficient (Impact on Attrition)")
plt.title("Feature Importance (Logistic Regression Coefficients)")
plt.gca().invert_yaxis()
plt.show()
```
